# **Steps to clean the data**


**Lowering** **text**

```python
df['text_lower']= df['text'].apply(lambda x: x.lower())
df
```

**Sentence Count**

```python
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
def tokenization(text_lower):
  tokens = sent_tokenize(text_lower)
  freq=len(tokens)
  return freq
#applying function to the column
df['sent_count']= df['text_lower'].apply(lambda x: tokenization(x))
df
```

**Sentence** **Tokenization**

```python
def tokenization(text_lower):
    tokens = sent_tokenize(text_lower)
    return tokens
#applying function to the column
df['sent_tokenied']= df['text_lower'].apply(lambda x: tokenization(x))
df
```

**Punctuation Removal:**

```python
import string
string.punctuation
def remove_punctuation(text_lower):
    punctuationfree="".join([i for i in text_lower if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
df['clean_text']= df['text_lower'].apply(lambda x:remove_punctuation(x))
df
```

 **Word Tokenization**

```python
from nltk.tokenize import word_tokenize
def tokenization(text):
    tokens = word_tokenize(text)
    return tokens
#applying function to the column
df['text_tokenied']= df['clean_text'].apply(lambda x: tokenization(x))
df
```

**Word Count**

```python
def tokenization(clean_text):
  tokens = word_tokenize(clean_text)
  freq=len(tokens)
  return freq
#applying function to the column
df['word_count']= df['clean_text'].apply(lambda x: tokenization(x))
df
```

**Stop Word Removal**

```python
nltk.download('stopwords')
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
df['no_stopwords']= df['text_tokenied'].apply(lambda x:remove_stopwords(x))
df
```

**Stemming**

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text
```

```python
df['text_stemmed']=df['no_stopwords'].apply(lambda x: stemming(x))
df
```

**Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
```

```python
nltk.download('omw-1.4')
df['text_lemmatized']=df['no_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
from collections import defaultdict
for j in range(3):
  # initializing Matrix
  test_list = df['no_stopwords'][j]
  
  # printing original list
  print("The original list is : " + str(test_list))
  
  temp = defaultdict(int)
  
  # memoizing count
  for sub in test_list:
      for wrd in sub.split():
          temp[wrd] += 1
  
  # getting max frequency
  res = max(temp, key=temp.get)
  
  # printing result
  print("Word with maximum frequency : " + str(res))
```

**`Word Count Visualisation`**

```python
from nltk import tokenize
import seaborn as sns
token_space = tokenize.WhitespaceTokenizer()

def counter(text, column_text, quantity):
    all_words = ' '.join([text for text in text[column_text]])
    token_phrase = token_space.tokenize(all_words)
    frequency = nltk.FreqDist(token_phrase)
    df_frequency = pd.DataFrame({"Word": list(frequency.keys()),
                                   "Frequency": list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns = "Frequency", n = quantity)
    plt.figure(figsize=(20,4))
    ax = sns.barplot(data = df_frequency, x = "Word", y = "Frequency", color = 'green')
    ax.set(ylabel = "Count")
    plt.xticks(rotation='vertical')
    plt.show()
```
